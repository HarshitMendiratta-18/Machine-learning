```python
# group imdb_df by movie genres
genre_groups = imdb_df.groupby('genre')
```

```python
# get crime movies group
# .get_group() lets you see one bucket
crime_movies = genre_groups.get_group('Crime')
print(crime_movies.head())
```

```python
# get mean of movie durations for each group
# We select the 'duration' column from the groups and calculate mean
print(genre_groups['duration'].mean())
```

```python
# change duration of all movies in a particular genre to mean duration of the group
# .transform() applies a function to the group but keeps the original shape (rows)
imdb_df['new_duration'] = genre_groups['duration'].transform(lambda x: x.mean())

print(imdb_df[['title', 'genre', 'duration', 'new_duration']].head())
```

```python
# drop groups/genres that do not have average movie duration greater than 120.
# .filter() keeps only  the rows belonging to groups that pass the test
long_duration_genres = genre_groups.filter(lambda x: x['duration'].mean() > 120)

print(long_duration_genres['genre'].unique())
```

```python
# group weekday wise bikers count
weekday_groups = bikes_df.groupby('weekday')

# get weekday wise biker count (Summing up all cyclists for Mondays, Tuesdays, etc.)
weekday_counts = weekday_groups.sum()
```

```python
# plot weekday wise biker count for 'Berri1'
# We probably want to reorder them so Monday comes first (optional but nice)
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# .reindex() is a pandas trick to sort by a custom list
weekday_counts['Berri1'].reindex(days).plot(kind='bar')
```
