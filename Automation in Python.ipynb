# Send Email in Python


```python
import smtplib
```

```python
import getpass
```

```python
from email.mime.text import MIMEText
```

```python
def send_email():
    sender_address='harshitmendiratta41@gmail.com'
    password=getpass.getpass()
    subject='Machine Learning'
    msg=''' Hello, I am Learning Machine Learning'''
    # server initialisation
    server= smtplib.SMTP('smtp.gmail.com',587)
    server.starttls()
    server.login(sender_address,password)

    #draft my message body
    msg=MIMEText(msg)
    msg['Subject']=subject
    msg['From']=sender_address
    msg['To']='ayurai6387@gmail.com'
    msg.set_param('importance','high value')
    #if we want to send it to multiple person then pass the array
    recipients='ayurai6387@gmail.com'
    server.sendmail(sender_address,recipients,msg.as_string())
send_email()
```
